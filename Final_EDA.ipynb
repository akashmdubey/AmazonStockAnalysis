```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
jpm = pd.read_csv("JPM.csv")
```

```python
jpm.shape
```

```python
jpm
```

## GS

```python
gs = pd.read_csv('GS.csv')
gs.head()
```

```python
gs = gs.rename(columns={'Open' : 'GS_Open', 'High' : 'GS_High', \
                        'Low' : 'GS_Low', 'Close' : 'GS_Close', \
                        'Adj Close' : 'GS_Adj_Close', \
                        'Volume' : 'GS_Volume' })
gs
```

## WFC

```python
wfc = pd.read_csv('WFC.csv')
wfc.head()
```

```python
wfc = wfc.rename(columns={'Open' : 'WFC_Open', 'High' : 'WFC_High', \
                        'Low' : 'WFC_Low', 'Close' : 'WFC_Close', \
                        'Adj Close' : 'WFC_Adj_Close', \
                        'Volume' : 'WFC_Volume' })
wfc
```

```python
jpm['WFC_Open'] = wfc['WFC_Open']
jpm['WFC_High'] = wfc['WFC_High']
jpm['WFC_Low'] = wfc['WFC_Low']
jpm['WFC_Close'] = wfc['WFC_Close']
jpm['WFC_Adj_Close'] = wfc['WFC_Adj_Close']
jpm['WFC_Volume'] = wfc['WFC_Volume']


jpm['GS_Open'] = gs['GS_Open']
jpm['GS_High'] = gs['GS_High']
jpm['GS_Low'] = gs['GS_Low']
jpm['GS_Close'] = gs['GS_Close']
jpm['GS_Adj_Close'] = gs['GS_Adj_Close']
jpm['GS_Volume'] = gs['GS_Volume']
jpm.head()
```

```python
jpm['yest_close'] = jpm['Adj Close'].shift(1)
jpm = jpm.fillna(55.047691)
jpm.head()
```

```python
jpm['WFC_Yest_Close'] = jpm['WFC_Adj_Close'].shift(1)
jpm = jpm.fillna(41.832123)
jpm.head()
```

```python
jpm['GS_Yest_Close'] = jpm['GS_Adj_Close'].shift(1)
jpm = jpm.fillna(144.872345)
jpm.head()
```

```python
jpm['Change_Percent'] = (jpm['Adj Close'] - jpm['yest_close']) / jpm['yest_close']
jpm.head()
```

```python
jpm['WFC_Change_Percent'] = (jpm['WFC_Adj_Close'] - jpm['WFC_Yest_Close']) / jpm['WFC_Yest_Close']
jpm['GS_Change_Percent'] = (jpm['GS_Adj_Close'] - jpm['GS_Yest_Close']) / jpm['GS_Yest_Close']
jpm.head()
```

```python
jpm['WFC_Change_Percent'] = jpm['WFC_Change_Percent'] * 100
jpm['GS_Change_Percent'] = jpm['GS_Change_Percent'] * 100
jpm.head()
```

```python
jpm['Change_Percent'] = jpm['Change_Percent'] * 100
jpm.head()
```

```python
jpm = jpm.drop(['Change_percent'], axis = 1)
jpm.head()
```

```python
jpm.info()
```

```python
jpm['Date'] = pd.to_datetime(jpm['Date'])
```

```python
jpm.info()
```

```python
jpm.head()
```

```python
plt.rcParams['figure.figsize'] = 30, 15
```

```python
cp1 = sns.lineplot(data = jpm, x = "Date", y = "Change_Percent", size = 10, palette = ['red'])
```

## CHANGE PERCENT

```python
jpm1 = jpm[jpm.Date > '2016-05-16']
jpm1
```

```python
plt.style.use('fivethirtyeight')
```

```python
date = jpm1['Date']
jpm_CP = jpm1['Change_Percent']
wfc_cp = jpm1['WFC_Change_Percent']
gs_cp = jpm1['GS_Change_Percent']
plt.plot(date, gs_cp, c = 'g', linewidth = 1, ls = '-', marker = 'v', ms = 3, label = 'GS')
plt.plot(date, jpm_CP, c = 'r', linewidth = 1, ls = '-', marker = 'o', ms = 3, label = 'JPM')
plt.plot(date, wfc_cp, c = 'b', linewidth = 1, ls = '-', marker = '^', ms = 3, label = 'WFC')

plt.xlabel('Date', fontsize = 18)
plt.ylabel('Stock Change (%)', fontsize = 18)
plt.legend(frameon=True, fancybox=True, shadow=True, \
           framealpha=1, prop = {'size':20}, )

plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.grid(color = 'gray', linestyle = '-', linewidth = 2)
plt.title("TREND FOR CHANGE IN STOCK PRICE (%)", fontsize=35, \
         color ='BLACK')
plt.savefig('cp.png')
plt.show()
```

## Volume

```python
date = jpm1['Date']
jpm_v = jpm1['Volume']
wfc_v = jpm1['WFC_Volume']
gs_v = jpm1['GS_Volume']
plt.plot(date, jpm_v, c = 'r', linewidth = 1, ls = '-', marker = 'o', ms = 3, label = 'JPM')
plt.plot(date, wfc_v, c = 'b', linewidth = 1, ls = '-', marker = '^', ms = 3, label = 'WFC')
plt.plot(date, gs_v, c = 'g', linewidth = 1, ls = '-', marker = 'v', ms = 3, label = 'GS')
plt.xlabel('Date', fontsize = 20)
plt.ylabel('Daily Volume', fontsize = 20)
plt.legend(frameon=True, fancybox=True, shadow=True, \
           framealpha=1, prop = {'size':20})

plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.grid(color = 'gray', linestyle = '-', linewidth = 2)
plt.title("COMPARISION OF DAILY TRADE VOLUME", fontsize=35, \
         color ='BLACK')
plt.savefig('v.png')
plt.show()
```

## OPENING - CLOSING PRICE

```python
date = jpm1['Date']
jpm_OC = ((jpm1['Open'] - jpm1['Close'])/jpm1['Open']) * 100
wfc_OC = ((jpm1['WFC_Open'] - jpm1['WFC_Close'])/jpm1['WFC_Open']) * 100
gs_OC = ((jpm1['GS_Open'] - jpm1['GS_Close'])/jpm1['GS_Open']) * 100

plt.plot(date, gs_OC, c = 'g', linewidth = 1, ls = '-', marker = 'v', ms = 3, label = 'GS')
plt.plot(date, wfc_OC, c = 'b', linewidth = 1, ls = '-', marker = '^', ms = 3, label = 'WFC')
plt.plot(date, jpm_OC, c = 'r', linewidth = 1, ls = '-', marker = 'o', ms = 3, label = 'JPM')

plt.xlabel('Date', fontsize = 20)
plt.ylabel('Opening - Closing Price (%)', fontsize = 20)
plt.legend(loc=0, bbox_to_anchor=(1,1), fontsize = 18)
plt.legend(frameon=True, fancybox=True, shadow=True, \
           framealpha=1, prop = {'size':20})
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.grid(color = 'gray', linestyle = '-', linewidth = 2)
plt.title("DAILY OPENING - CLOSING PRICE", fontsize=35, \
         color ='BLACK')
plt.savefig('oc.png')
plt.show()
```
